### Objetivos de la práctica:

El objetivo principal de esta práctica es predecir las ventas de Apple en lo que queda de año 2021 y el próximo 2022 mediante la aplicación de los modelos ETS y ARIMA, compararlos y, posteriormente, responder a una serie de preguntas las cuales nos ayudará a realizar una evaluación acerca de las series.  

Los datos para poder realizar la prácticas son facilitados por la plataforma Bloomberg, y son datos pasados del 1º trimestre de 1990 hasta el 2º trimestre del vigente año.

### Ventas de Apple

Las ventas de Apple consta de dos fases que se aprecia con bastante claridad en la gráfica de abajo: 

- Una fase desde su creación hasta el 2007/2008 donde se ve un valle, es decir, la empresa tenía una tendencia constante de ventas.

- Otra fase con una tendencia alcista a partir del 2008. 

El motivo de este hecho es debido a la introducción de los dispositivos móviles inteligentes de Apple, el iPhone, que es el producto que da el mayor sustento económico a la empresa de Cupertino. Además, con el paso del tiempo y debido al nivel alto de innovación de la marca, han ido introduciendo nuevos productos con el fin de captar nuevos clientes, como son las tablets iPad, los portátiles MacBook o los auriculares inalámbricos AirPods, entre otros.

### Estacionariedad de las ventas

Una serie temporal es estacionaria si su media y su varianza se mantienen constantes con el paso del tiempo. En el caso de que no sea estacionaria, la serie temporal puede mostrar una tendencia o cambios al alza o a la baja en la media y/o en la varianza, y pueden presentar efectos estacionales que sean similares en el tiempo. 

Dicho esto, para comprobar la estacionariedad de las ventas, se realiza el cálculo de las medias y varianzas y del comportamiento estacional y se grafica:

Como se puede apreciar visualmente, Apple presenta una tendencia alcista en el tiempo en las medias y varianzas, y el comportamiento estacional, el 1er trimestre fiscal (que sería el 4º trimestre en año natural) tiene más ventas que el resto del año. El motivo de esto es porque Apple se reserva en los meses de septiembre-octubre para la presentación de sus nuevos productos y porque en las fechas previas y durante el periodo navideño (como el Black Friday), Apple "sufre" un incremento de ventas.

### Modelos de Predicción: ETS vs ARIMA

Observando la gráfica posterior, los modelos de predicción no es que varíen demasiado, pues son bastante similares entre sí. ### Efecto COVID en las ventas de Apple

En el siguiente Data Frame que se ha creado para poder valorar las estimaciones y diferencia, se puede observar que las estimaciones del modelo han estado, por lo general, por debajo de los valores reales. El motivo de esto esto podría deberse a la volatilidad y los cambios acentuados por el periodo de la pandemia por COVID-19, donde Apple ha conseguido incluso incrementar sus ventas en un periodo tan delicado como ese. 

La única diferencia porcentual donde el modelo supera al número de ventas de los datos reales es en el segundo trimestre del 2020, momento en el que comienza la pandemia. Aun así, no es una diferencia abismal, incluso más adelante, a partir del vigente año, las ventas de Apple ha llegado a ser un 40% más de lo esperado en el modelo.

### Comparativa con los datos predictivos de Barclays

El posterior Data Frame se realiza una comparativa de las predicciones para finales del 2021 y 2022 del modelo ARIMA realizado anteriormente con las predicciones para el mismo periodo realizado por Barclays.

Obviando la coincidencia del año 2020 (coinciden porque son datos pasados), para el año 2021 se espera por parte de ARIMA que las ventas sean de ≈373.000 mill USD, mientras que Barclays predice unos 368.000 mill USD, es decir, un 0'92% menos, lo cual no es significativo, aparte que el porcentaje de crecimiento de las ventas en ambos modelos están en torno al 34-35%, son bastante coincidentes.

Sin embargo, de cara al año 2022, los datos de Barclays predice unas ventas de ≈379.000 mill USD y un crecimiento del 2'62% mientras que el modelo ARIMA predice casi 438.000 mill USD, lo que supone un crecimiento del 17'62%. Esto significa una diferencia del modelo ARIMA con el modelo de Barclays de un 13'55%, lo cual ya hace sea significativo y que se cuestione cómo calculó Barclays su predicción.

## Conclusiones finales al análisis y predicción

- Apple posee unas ventas con un crecimiento y una tendencia positiva. Sus medias y varianzas varían con el tiempo, lo que hace ver que tiene un comportamiento no estacionario, y la serie temporal presenta estacionalidad y un crecimiento de ventas más alto en su primer trimestre fiscal.


- Los modelos usados para esta práctica para la predicción de las ventas de Apple poseen un porcentaje de MAPE alto, lo que concluimos que estos modelos no son del todo fiables, aunque ARIMA ha sido la que más se acercó a los datos reales.


- Apple, a diferencia de otras empresas, ha conseguido no solo no tener pérdidas sino tener un claro crecimiento fuera de las expectativas previstas durante la pandemia del COVID-19.


- El resultado del crecimiento de las ventas de Apple para el año 2022, basado en la predicción Barclays, es del 2'62%, lo que denota un carácter conservador. Esto podría ser que Barclays puede conocer mejor el modelo de negocio de la empresa y de ahí que su predicción no sea muy alta.
